```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.set_printoptions(threshold=np.inf, suppress=True, precision=8)
```

# Membuat sinyal diskrit

Pertama-tama kita akan membuat sinyal diskrit terlebih dahulu untuk kita tampilkan plotting dari sinyal tersebut

## Sinyal A

Sinyal ini memiliki karakteristik sebagai berikut:



```python
n = np.arange(0, 30 + 1) # Membuat array dari angka 0 s/d 30

sinyal_a = np.array(np.sin(0.5 * np.pi * n), dtype=np.float128)

print(sinyal_a)

plt.stem(n, sinyal_a, label="Sinyal A")

plt.title("Sinyal A = sin(0.5 * PI * n)")
plt.xlabel("Sample ke-n")
plt.ylabel("Magnitudo")
plt.grid(True)
plt.show()
```

## Sinyal B
Sinyal ini memiliki karakteristik melonjak dari 0 ke 1

Karena ketika `n` masih di bawah 5 nilai nya akan 0, sebaliknya jika nilai `n` sama dengan atau lebih besar dari 5 maka nilainya akan menjadi 1

```python
sinyal_b = np.where(n >= 5, 1, 0)

plt.plot(sinyal_b, label="Sinyal B")
plt.title("Sinyal B = u[n-5]")
plt.legend()
plt.grid(True)
plt.show()
```
