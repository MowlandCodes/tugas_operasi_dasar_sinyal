# Membuat sinyal diskrit

Pertama-tama kita akan membuat sinyal diskrit terlebih dahulu untuk kita tampilkan plotting dari sinyal tersebut

## Sinyal A $x[n] = \sin(0.5\pi n)$

Sinyal ini memiliki karakteristik sebagai berikut:

- **Sifat Periodisitas (Periodik)**: Berbeda dengan sinyal pertama, sinyal sinusoidal ini bersifat periodik karena nilainya berulang secara konstan setelah mencapai interval tertentu.
- **Periode Fundamental ($N$)**: Sinyal ini memiliki periode fundamental sebesar $N = 4$ sampel.

  **Perhitungan Matematis**:

  Frekuensi sudut digital $\omega_0 = 0.5\pi$. Berdasarkan rumus $\omega_0 = \frac{2\pi m}{N}$, maka:
  $$0.5\pi = \frac{2\pi m}{N} \implies N = \frac{2\pi m}{0.5\pi} = 4m$$
  Untuk nilai konstanta bulat terkecil ($m = 1$), diperoleh periode fundamental $N = 4$.

- **Nilai Amplitudo Terbatas (Bounded)**: Amplitudo sinyal ini terbatas dalam rentang nilai $[-1, 1]$. Berdasarkan urutan sampelnya, sinyal ini membentuk pola sekuensial yang berulang setiap 4 sampel, yaitu: $[0, 1, 0, -1]$.

```python
n = np.arange(0, 30 + 1) # Membuat array dari angka 0 s/d 30

sinyal_a = np.array(np.sin(0.5 * np.pi * n), dtype=np.float128)

plt.stem(n, sinyal_a, label="Sinyal A")

plt.title("Sinyal A = sin(0.5 * PI * n)")
plt.xlabel("Sample ke-n")
plt.ylabel("Magnitudo")
plt.grid(True)
plt.show()
```

## Sinyal B $x_2[n] = u[n-5]$
Sinyal ini memiliki karakteristik sebagai berikut:

- **Sifat Periodisitas (Aperiodik)**: Sinyal ini bersifat aperiodik (tidak periodik) karena tidak memiliki pola nilai yang berulang dalam interval waktu tertentu. Sinyal hanya mengalami satu kali perubahan kondisi (state).
- **Kausalitas (Sinyal Kausal)**: Sinyal ini termasuk dalam kategori sinyal kausal. Secara definisi, sinyal kausal bernilai nol untuk rentang waktu negatif ($n < 0$). Pada kasus ini, sinyal baru mulai bernilai pada $n \ge 5$.
- **Nilai Amplitudo**: Amplitudo sinyal bersifat deterministik biner, di mana nilai amplitudo hanya terdiri dari dua kondisi: bernilai $0$ untuk fase non-aktif ($n < 5$) dan bernilai $1$ untuk fase aktif ($n \ge 5$).

```python
sinyal_b = np.where(n >= 5, 1, 0)

plt.stem(sinyal_b, label="Sinyal B")
plt.title("Sinyal B = u[n-5]")
plt.grid(True)
plt.show()
```
